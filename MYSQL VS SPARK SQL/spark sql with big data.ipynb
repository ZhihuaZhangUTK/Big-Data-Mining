```python
def run_time(filename):
    df = spark.read.csv(filename, header=True)
    df.createOrReplaceTempView("hotel")
    %time sqlContext.sql("SELECT * FROM hotel WHERE srch_ci = '2014-05-04' AND srch_co='2014-05-07'").show()
    %time sqlContext.sql("SELECT * FROM hotel WHERE srch_ci in ('2014-05-04', '2014-05-07', '2014-05-10')").show()
    %time sqlContext.sql("Select COUNT(*) from hotel").show()
    %time sqlContext.sql("SELECT DISTINCT user_id FROM hotel").show()
    %time sqlContext.sql("SELECT is_mobile, COUNT(*) FROM hotel GROUP BY is_mobile").show()
    %time sqlContext.sql("SELECT user_id, orig_destination_distance FROM hotel WHERE srch_ci = '2014-05-04' AND srch_co='2014-05-07' ORDER BY orig_destination_distance").show()
    %time sqlContext.sql("SELECT user_location_country, AVG(orig_destination_distance) FROM hotel GROUP BY user_location_country").show()
```
